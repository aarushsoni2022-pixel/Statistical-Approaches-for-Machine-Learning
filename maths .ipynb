```python
data=[2,5,5,5,5,6,6,7]
freq ={}
for x in data:
    freq[x]=freq.get(x,0)+1
    
max_freq = max(freq.values())

modes=[]
for k,v in freq.items():
    if v==max_freq:
        modes.append(k)
print("dataset mode:", modes)
```

```python
#standard devation

dataset=[48,49,50,51,52]
n=len(dataset)
dataset_mean = sum(dataset)/len(dataset)
sum_sqrd=0
for x in dataset:
    sum_sqrd+=(x-dataset_mean)**2
std_dev =(sum_sqrd/n)**0.5
print("standard deviation",std_dev)
var = sum_sqrd/n
print("variance",var)
```

```python
#quartiles of given data set 
data =[20,22,23,25,30,32,36]
data.sort()
n=len(data)
#median (q2)
if n%2==0:
    mid1=data[n//2-1]
    mid2=data[n//2]
    q2=(mid1+mid2)/2
else:
    q2=data[n//2]
#lower and upper halves 
if n%2==0:
    lower_half=data[:n//2]
    upper_half=data[:n//2]
else:
    lower_half=data[:n//2]
    upper_half=data[n//2+1:]
#q1 (median of lower half)
n1 = len(lower_half)
if n1%2==0:
    q1=(lower_half[n1//2-1]+lower_half[n1//2])/2
else:
    q1=lower_half[n1//2]
#q3 (median of  upper half)
n2=len(upper_half)
if n2%2==0:
    q3=(upper_half[n2//2-1]+upper_half[n2//2])/2
else:
    q3=upper_half[n2//2]


print(q1)
print(q2)
print(q3)
```
